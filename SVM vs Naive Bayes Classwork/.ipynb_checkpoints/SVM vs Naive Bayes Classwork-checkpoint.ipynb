# Natural Language Processing
#### start NLP Processissing

We will use TF-IDF to convert the text reviews into numerical data.

we need to tokenize the text and clean it by removing stopwords, punctuation, and performing stemming:

- Tokenization.

في هاي الخطوة هنحول النصوص لكلمات

```python
import nltk
from nltk.tokenize import wordpunct_tokenize
nltk.download('punkt')
```

```python
reviews = df['Review'].values
tokenize_phase =[wordpunct_tokenize(i.lower()) for i in reviews]
reviews[0] , tokenize_phase[0]
```

-----------------------------------------------------------------------------------------------------

- Stopwords and punctuation removal

في هاي الخطوة هنعمل كلين للداتا ونقيم علامات الترقيم والكلمات اللي ما الها تاثير ولا دلالة في المعنى وممكن تعمل تحيز في الموديل اللي هنبني


```python
nltk.download('stopwords')
en_stopwords = nltk.corpus.stopwords.words('english')
```

```python
cleaning_phase1 =[]
for reviews_tokens in tokenize_phase:
  cleaning_phase1.append([i for i in reviews_tokens if i not in en_stopwords])

tokenize_phase[0] , cleaning_phase1[0]
```

```python
to_remove = string.punctuation+'0123456789'
cleaning_phase2=[]
for reviews_tokens in cleaning_phase1:
  cleaning_phase2.append([i for i in reviews_tokens if i not in to_remove])

cleaning_phase1[0] , cleaning_phase2[0]
```

-----------------------------------------------------------------------------------------------------

- Stemming (Use Snowball stemmer)

في هاي الخطوة هنرجع كل كلمة لأصلها أو جذرها بحيث نجمع الكلمات المتشابهة وان كانت مش نتائج دقيقية لكن عملية التحويل واحدة للكلمات المتشابهة

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer(language='english')
```

```python
stemming_phase=[]
for reviews_tokens in cleaning_phase2:
  stemming_phase.append([stemmer.stem(i) for i in reviews_tokens])

cleaning_phase2[0] , stemming_phase[0]
```

-----------------------------------------------------------------------------------------------------

- Vectorization by using TF-IDF.

في هاي الخطوة هنحول النصوص لفيكتورز وارقام قادر الموديل يتعامل معها

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Initialize TF-IDF vectorizer
vectorizer = TfidfVectorizer()
```

```python
# make string for vectorization
vectorize_phase=[' '.join(i) for i in stemming_phase]
vectorize_phase[0]
```

```python
# Fit the vectorizer and transform the reviews into TF-IDF features
tfidf_matrix = vectorizer.fit_transform(vectorize_phase) # ready to use in ML

tfidf_matrix.toarray()
```

**Rows: Each row corresponds to a different review.**

**Columns: Each column corresponds to a different term from the vocabulary.**

**Values: Each value represents the TF-IDF score for a specific term in a specific document.**

#############################################

**Zero Values: A value of 0 means the term does not appear in the document.**

**Non-Zero Values: A non-zero value indicates the importance of the term in the document based on its frequency and how unique it is across the entire corpus.**

```python
feature_names = vectorizer.get_feature_names_out()
feature_names
```

```python
# words weghit on first row
for i, weight in enumerate(tfidf_matrix.toarray()[0]):
    print(f"{feature_names[i]}: {weight}")
```

```python
# words weghit on first row
for i, weight in enumerate(tfidf_matrix.toarray()[0]):
  if weight != 0:
    print(f"{feature_names[i]}: {weight}")
```

   - **Explanation**: The term "delici has a TF-IDF weight of 0.53. This relatively high value suggests that the term is quite important in the context of this document. It indicates that the term "delici" is used frequently within the document and is not as common across the entire corpus.
   - **Importance**: This high weight shows that "delici" is a key term in describing the positive attributes of the food in this review.

# Train SVM Models

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, df['Label'], test_size=0.2, random_state=42)
```

```python
# Train SVM
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)

# Evaluate models
svm_accuracy = accuracy_score(y_test, svm_predictions)

print(f"SVM Accuracy: {svm_accuracy}")
```

```python
# Train Naïve Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
nb_predictions = nb_model.predict(X_test)

nb_accuracy = accuracy_score(y_test, nb_predictions)

print(f"Naïve Bayes Accuracy: {nb_accuracy}")
```

```python
# Plotting the accuracy comparison
models = ['SVM', 'Naïve Bayes']
accuracies = [svm_accuracy, nb_accuracy]

plt.bar(models, accuracies, color=['blue', 'orange'])
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison of SVM and Naïve Bayes')
plt.ylim(0, 1)
plt.show()
```

```python
import pandas as pd
import pickle
from sklearn.metrics import accuracy_score

# New test reviews
new_reviews = [
    "Absolutely loved the food and the service was fantastic!",
    "The waiter forgot our order. What a great experience!",
    "The pizza was so good, I could eat it every day.",
    "I really enjoyed the ambiance, but the food was awful.",
    "Best service ever. Waited an hour for cold food."
]

# Load the TF-IDF vectorizer
with open('tfidf_vectorizer.pkl', 'rb') as file:
    vectorizer_loaded = pickle.load(file)

# Vectorize the new reviews using the loaded vectorizer
new_reviews_vectorized = vectorizer_loaded.transform(new_reviews)

# Make predictions with the new reviews
svm_test_predictions = svm_model.predict(new_reviews_vectorized)
nb_test_predictions = nb_model.predict(new_reviews_vectorized)

# Create a DataFrame to display the results
results = pd.DataFrame({
    "Review": new_reviews,
    "Expected": [1, 0, 1, 0, 0],
    "SVM Prediction": svm_test_predictions,
    "Naïve Bayes Prediction": nb_test_predictions
})

# Compute accuracy
svm_accuracy = accuracy_score(results["Expected"], results["SVM Prediction"])
nb_accuracy = accuracy_score(results["Expected"], results["Naïve Bayes Prediction"])

print(f"SVM Accuracy: {svm_accuracy}")
print(f"Naïve Bayes Accuracy: {nb_accuracy}")

# Display the results
results
```

### Final Discussion:

**Which model had higher accuracy?**
  - The SVM model had higher accuracy (0.6) compared to the Naïve Bayes model (0.4) for these test reviews.

    **Which model was faster?**
  - The Naïve Bayes model was generally faster to train and predict due to its simplicity.

**When would SVM be better than Naïve Bayes and vice versa?**
  - **SVM**: Better for high-dimensional spaces and non-linearly separable data, useful when you need robust results with smaller datasets.
  - **Naïve Bayes**: Suitable for large datasets with many features, works well when the independence assumption is approximately true, and faster training and prediction times make it ideal for real-time applications.

# Scraping comments

```python
import requests # use it for send request to page url that i want scrapping data from
from bs4 import BeautifulSoup  # use it in web scraping process 'parsing HTML content'
import time # add delays between requests to avoid blocke
import random  # use it to add delays between requests "to not blocked from site"
from fake_useragent import UserAgent  # used to to generate random user agents for each request .. same user agent get 503 can send req only one
import re  # used for regular expressions -  some data shoud extacted from specif expressions
```

```python
# get the soup object "to parse" from the URL
def get_soup(url, headers):
    for _ in range(5):  # Retry up to 5 times if the request fails
        try:
            response = requests.get(url, headers=headers)  # sedn GET HTTP request to the URL
            response.raise_for_status()  # raise an HTTP Error for bad responses 400 and 500 "error has occurred during the process"
            return BeautifulSoup(response.content, 'html.parser')  # parse HTML content to use it and extract data
        except requests.RequestException as e:
            # print(f"Failed to retrieve page {url}, error: {e}")  # error message
            time.sleep(random.randint(1, 5))  # Wwit for a random time between 1 to 5 seconds before retrying
    return None  # Return None if the request fails after 5 retries
```

```python
def parse_review(review, page_number):
    def safe_find(element, search_dict, text=False):
        try:
            found = element.find(**search_dict)
            return found.text.strip() if (found and text) else found
        except AttributeError:
            return None

    # Extract review title
    title = safe_find(review, {'name': 'a', 'attrs': {'class': 'BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS'}}, text=True)

    # Extract review text
    text = safe_find(review, {'name': 'span', 'attrs': {'class': 'JguWG'}}, text=True)

    # Return review details
    return {
        'Title': title,
        'Review': text
    }
```

```python
def scrape_page(url, headers, page_number):
    soup = get_soup(url, headers)
    if not soup:
        return []

    # Extract review divs
    review_divs = soup.find_all('div', {'class': '_c'})

    # Parse each review
    reviews = [parse_review(review, page_number) for review in review_divs]

    return [review for review in reviews if review]
```

```python
ua = UserAgent()
headers = {
    "accept-language": "en-US,en;q=0.9",
    "accept-encoding": "gzip, deflate, br",
    "User-Agent": ua.random,
    "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "Connection": "close"
}

base_url = 'https://www.tripadvisor.com/Restaurant_Review-g295419-d12081579-Reviews-or{page_number}-Ginger_Leaf-Jeddah_Makkah_Province.html'
all_reviews = []
page_numbers = [0, 15, 30, 45, 60] 

for page_number in page_numbers:
    headers['User-Agent'] = ua.random  # New random User-Agent for each request
    url = base_url.format(page_number=page_number)
    print(url)
    reviews = scrape_page(url, headers, page_number)

    if not reviews:
        break

    all_reviews.extend(reviews)  # Add the reviews to the list

    # Delay to avoid being blocked by the site
    time.sleep(random.randint(1, 5))

# Convert all reviews to a DataFrame
df = pd.DataFrame(all_reviews)
df.index += 1  # Index from 1
df.to_csv('tripadvisor_reviews.csv', index_label='index')  # Save the DataFrame to a CSV file
```
