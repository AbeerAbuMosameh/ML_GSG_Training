## Import needed Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

## Make Dataset

```python
data = {
    'size': [1200, 1500, 800, 1000, 2000],
    'bedrooms': [3, 4, 2, 2, 5],
    'age': [10, 5, 20, 15, 8],
    'distance': [5, 3, 10, 8, 2],
    'price': [250000, 320000, 180000, 210000, 450000]
}
dataset = pd.DataFrame(data)
dataset
```

## Features and labels


```python
X = dataset[['size', 'bedrooms', 'age', 'distance']]
y = dataset['price']
```

## Regression model


```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

# Plot Tree

```python
# Visualizing the decision tree
plt.figure(figsize=(12, 8))
plot_tree(regressor, feature_names=X.columns, filled=True, rounded=True)
plt.title("Decision Tree - Real Estate Prices (Regression)")
plt.show()
```

# Predict the price for a house with given features

```python
predicted_price = regressor.predict([[1400, 3, 12, 6]])
print(f"Predicted price for the house: ${predicted_price[0]:,.2f}")
```

# Visualize the decision tree regression for size

```python
X_size = dataset[['size']]
regressor_size = DecisionTreeRegressor(random_state=0)
regressor_size.fit(X_size, y)

X_grid = np.arange(min(X_size.values), max(X_size.values), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X_size, y, color='red')
plt.plot(X_grid, regressor_size.predict(X_grid), color='blue')
plt.title('House Prices vs Size (Decision Tree Regression)')
plt.xlabel('Size')
plt.ylabel('Price')
plt.show()
```

# Export the decision tree to a .dot file

```python
export_graphviz(regressor, out_file='tree.dot', 
               feature_names=['size', 'bedrooms', 'age', 'distance'],
               filled=True, rounded=True, special_characters=True)
```

# Classification Model

```python
# make dataset , split Features and label to predict it
data = {
    'size': [1200, 1500, 800, 1000, 2000],
    'bedrooms': [3, 4, 2, 2, 5],
    'age': [10, 5, 20, 15, 8],
    'distance': [5, 3, 10, 8, 2],
    'price_category': ['medium', 'high', 'low', 'medium', 'high'] # add for classification model -  classify houses based on other features to category
}

dataset = pd.DataFrame(data)
X = dataset[['size', 'bedrooms', 'age', 'distance']]
y = dataset['price_category']
```

```python
# Decision Tree Classifier model creation
classifier = DecisionTreeClassifier(random_state=0)
classifier.fit(X, y)
```

```python
# Visualizing the decision tree - to understand the classification process
plt.figure(figsize=(8, 8))
plot_tree(classifier, feature_names=X.columns, filled=True, rounded=True)
plt.title("Decision Tree - Real Estate Prices (Classification)")
plt.show()
```

```python
# predict a price category
predicted_category = classifier.predict([[1400, 3, 12, 6]])
print(f"Predicted price category for the house: {predicted_category[0]}")
```
